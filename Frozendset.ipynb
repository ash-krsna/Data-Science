```python
import numpy as np

#Create an object of ndarray from frozenset object Value
x = frozenset({10,20,30,40,50,60,70,80,90,15,25,35,55,65,75,85})
print("Val of x ={} and type = {}".format(x,type(x)))

a = np.array(x)

print("-"*50)
print("Content of ndarray = {} Type = {}".format(a,type(a)))
print("Dimension of a =",a.ndim)
print("D type of a =",a.dtype)
print("Size of a =",a.size)
print("Shape of a =",a.shape)
print("-"*50)
```
