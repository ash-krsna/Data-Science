```python
import numpy as np

# Normal Python list
list1 = [1, 2, 3, 4, 5]
print("Python List:", list1)

# Convert to NumPy array
arr1 = np.array(list1)
print("NumPy Array:", arr1)

# Check type difference
print("Type of list1:", type(list1))
print("Type of arr1:", type(arr1))
```

```python
print("Array from list:", np.array([1, 2, 3, 4, 5]))
print("Zeros array:\n", np.zeros((2, 3)))
print("Ones array:\n", np.ones((3, 2)))
print("Arange (like range):", np.arange(0, 10, 2))
print("Linspace (evenly spaced):", np.linspace(0, 1, 5))
```

```python
arr = np.array([[10, 20, 30], [40, 50, 60]])
print("Array:\n", arr)
print("Shape:", arr.shape)
print("Size:", arr.size)
print("Dimensions:", arr.ndim)
print("Data Type:", arr.dtype)
```

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0])      # First element
print(arr[-1])     # Last element
print(arr[1:4])    # From 2nd to 4th element
```
