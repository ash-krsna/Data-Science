```python
### creating array from list
```

```python
### List vs numpy arra
```

```python
py_list = [1,2,3]
print("Python list multiplication",py_list * 2)

np_array = np.array([1,2,3]) #element wise multiplication
print("Python list multiplication",np_array * 2)

import time
start = time.time()
py_list = [i*2 for i in range(1000)]
print("\n List Operation time:",time.time() - start)

start = time.time()
np_array = np.arange(1000) * 2
print("\nNumpy operation time:",time.time() - start)
```

```python
### Vector, Matrix and Tensor
```

```python
vector = np.array([1,2,3])
print("Vector:",vector)

matrix = np.array([[1,2,3],
                   [4,5,6]])
print("Matrix: ",matrix)

tensor = np.array([[[1,2],[3,4]],
                  [[5,6],[7,8]]])
print("Tensor array:\n ",tensor)
```

```python
### Array Properties
```

```python
arr = np.array([[1,2,3],
                 [4,5,6]])
print("shape :",arr.shape)
print("Dimension:", arr.ndim)
print("Size:",arr.size)
print("Datatype",arr.dtype)
```

```python
### Array Reshaping
```

```python
arr = np.arange(12)
print("Original array:",arr)

reshaped = arr.reshape((3,4))
print("\nReshaped array: \n",reshaped)

flattend = reshaped.flatten()
print("\n Flattened Array", flattend)
```

```python
# ravel (returns view, insted of copy)
raveled = reshaped.ravel()
print("\n raveled array : ",raveled)
```

```python
#Transpose
transpose = reshaped.T
print("\t Transposed array:",transpose)
```
